```python
from bs4 import BeautifulSoup
import re

# Скопируй функции из data_preprocessing.ipynb
def html_to_text(html):
    soup = BeautifulSoup(html, "html.parser")
    text = soup.get_text(separator=" ")
    text = text.replace("\xa0", " ")
    return text.strip()

def is_noise(text):
    if len(text) < 3:
        return True
    if re.match(r"^[\W\d_]+$", text):
        return True
    if re.match(r"^[a-zA-Z]+$", text):
        return True
    return False
```

```python
from datasets import Dataset

def preprocess_function(examples):
    inputs = tokenizer(
        examples['MessageText'],
        truncation=True,
        padding='max_length',
        max_length=128,
        return_tensors="pt"
    )

    return inputs

train_dataset = Dataset.from_pandas(data).map(preprocess_function, batched=True)
test_dataset = Dataset.from_pandas(test_data).map(preprocess_function, batched=True)
```

```python
import torch
import numpy as np
from sklearn.metrics import recall_score

# Переводим модель в режим оценки
model.eval()

# Подготовка данных
input_ids = torch.tensor(test_dataset['input_ids']).to(model.device)
attention_mask = torch.tensor(test_dataset['attention_mask']).to(model.device)
true_labels = torch.tensor(test_dataset['labels']).cpu().numpy()

# Получаем предсказания
with torch.no_grad():  # Отключаем вычисление градиентов
    outputs = model(input_ids=input_ids, attention_mask=attention_mask)
    logits = outputs.logits

# Преобразуем логиты в предсказанные классы
predicted_labels = torch.argmax(logits, dim=-1).cpu().numpy()

# Вычисляем recall
recall = recall_score(true_labels, predicted_labels, average=None)

# Вывод recall для каждого класса
print("Recall для каждого класса:")
for i, rec in enumerate(recall):
    print(f"Класс {i}: {rec:.2f}")

# Вывод macro-averaged recall
macro_recall = recall_score(true_labels, predicted_labels, average="macro")
print(f"\nMacro-averaged recall: {macro_recall:.2f}")

# Вывод weighted-averaged recall
weighted_recall = recall_score(true_labels, predicted_labels, average="weighted")
print(f"Weighted-averaged recall: {weighted_recall:.2f}")
```

```python
import torch
import asyncio
from IPython.display import display, clear_output
import ipywidgets as widgets
from torch import nn
from sklearn.metrics import recall_score
import random

# Определяем устройство
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Определяем функцию потерь и оптимизатор
# Рассчитать веса классов (предположим, класс 0 - редкий)
class_weights = torch.tensor([2.0, 1.0, 1.0], device=device)  # Усилить вес класса 0
criterion = nn.CrossEntropyLoss(weight=class_weights)
#
# Создаем интерактивные элементы
output = widgets.Output()
feedback_event = asyncio.Event()  # Событие для ожидания фидбэка
user_feedback = -1
```

```python
# Функция для ввода пользователя
def get_user_feedback():
    while True:
        try:
            feedback = int(input("Введите класс (0, 1, 2): "))
            if feedback in [0, 1, 2]:
                return feedback
            else:
                print("Некорректный ввод. Введите 0, 1 или 2.")
        except ValueError:
            print("Некорректный ввод. Введите число.")
```

```python
from sklearn.metrics import classification_report

async def async_rlhf_training(model, train_dataset, test_dataset, num_epochs=1, feedback_interval=5):
    model.train()
    display(output)

    optimizer = torch.optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=3e-6
    )

    random_indices = random.sample(range(len(train_dataset)), 20)
    random_samples = [train_dataset[i] for i in random_indices]

    final_recall = None
    total_feedback = 0  # Счетчик запрошенных фидбеков

    for i, batch in enumerate(random_samples):
        feedback_event.clear()

        text = batch["MessageText"]
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.set_grad_enabled(True):
            outputs = model(**inputs)

        probs = torch.softmax(outputs.logits, dim=-1)
        confidence, prediction = torch.max(probs, dim=-1)
        confidence = confidence.item()
        prediction = prediction.item()

        with output:
            clear_output(wait=True)
            print(f"Текст: {text}")
            print(f"Предсказание: {prediction} (Уверенность: {confidence:.2%})")

        user_feedback = None
        # Запрашиваем фидбэк только при низкой уверенности
        if confidence < 0.7 or (prediction == 0 and confidence < 0.6):
            def get_feedback():
                try:
                    fb = int(input("Введите класс (0-2) или -1 для пропуска: "))
                    return fb if fb in {0,1,2} else get_feedback()
                except:
                    return get_feedback()

            user_feedback = get_feedback()
            total_feedback += 1

        # Если фидбэк не получен и модель уверена - пропускаем пример
        if user_feedback is None:
            continue

        # Обучение только на примерах с фидбэком
        target = torch.tensor([user_feedback], device=device)
        loss = criterion(outputs.logits, target)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Оценка каждые N примеров с фидбэком
        if total_feedback % feedback_interval == 0 and total_feedback > 0:
            model.eval()
            all_labels, all_preds = [], []

            for test_batch in test_dataset:
                inputs = tokenizer(test_batch["MessageText"],
                                 return_tensors="pt",
                                 padding=True,
                                 truncation=True)
                inputs = {k: v.to(device) for k, v in inputs.items()}

                with torch.no_grad():
                    outputs = model(**inputs)

                preds = torch.argmax(outputs.logits, dim=-1).cpu().numpy()
                labels = test_batch["labels"]

                all_labels.extend([labels] if isinstance(labels, int) else labels.tolist())
                all_preds.extend(preds.tolist())

            recall = recall_score(all_labels, all_preds, average="macro")
            final_recall = recall

            with output:
                clear_output(wait=True)
                print(f"Прогресс: {total_feedback}/{len(random_samples)} примеров")
                print(f"Текущий Recall: {recall:.4f}")
            print(classification_report(all_labels, all_preds, target_names=["Class 0", "Class 1", "Class 2"]))

            model.train()

    return final_recall
```
